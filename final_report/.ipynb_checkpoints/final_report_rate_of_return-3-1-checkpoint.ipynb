## Methods and Results

---
### Cleaning and wrangling
- To reliminate the cumulative effects of pricing and other factors, we need to change the raw numeric figures to ROC (rate of change). 
- The formula of ROC given by:
$$
ROC = \frac{V_{final} - V_{initial}}{V_{initial}}\\
$$

- Therefore, we would first eleminate the non-numeric variable `date`.

```r
eth_selected <- eth_raw %>% select(-date)

eth_shift1 <- eth_selected %>% head(nrow(eth_raw) - 1)
eth_shift1 <- rbind(rep(0, ncol(eth_shift1)), eth_shift1)

eth <- ((eth_selected - eth_shift1) / eth_shift1) %>% tail(nrow(eth_selected) - 1)

head(eth, 2)
tail(eth, 2)
```

Observe that there are many `NaN`s and infinity values in the ROC form of our dataset. One approach to encounter this problem is to use `na.omit()`. However, when we apply this method to the whole dataframe, only few valid rows are left.

```r
cat('Number of rows left after removing invalid values:', na.omit(eth) %>% nrow())
```

So we have to remove variables that constribute the most invalid values. A much more comprehensive dataset is produced after these columns are excluded, which is helpful for our further analysis.

```r
# remove NaNs
eth <- eth %>% 
    select(-trade_vol_l, -trade_vol_x, -margin_call_l, -margin_call_x) %>%
    na.omit()

# remove infinity values
eth <- eth[is.finite(rowSums(eth)),]

head(eth, 2)
tail(eth, 2)
cat('Number of rows left after removing invalid values:', eth %>% nrow())
```

Next, we can derive our target response: **return rate of the next day**, by moving our `price` column one unit upward. THis gives our new column `TARGET_return_nextday`.

```r
# switch the next-day price as the response variable of the same-day variables
price_temp = pull(eth, price)

# we delete the first row of target because we don't have the explanatory variables from previous day
TARGET_return_nextday <- price_temp[-1]

# we delete the last row because we don't have the actual target variable of the latest day, we'll have to wait till monday!
eth_data <- eth[-length(price_temp),]
```

```r
# merge the operation with original dataset
eth_data <- cbind(eth_data, TARGET_return_nextday)
```

```r
cat('Number of inputs:', ncol(eth_data) - 1)
```

We now have:  
- `TARGET_return_nextday` as our **response variable**, and 45 variables as inputs to choose from.

### Association between explanotory variables and responses
To assure the viability of this model, we want to check that the collected data of the previous day are indeed associated with our targeted response, to do that, we first use`cor()` to test the correlation between our explanotory variables and response.
> Also notice: right now `TARGET_return_nextday` is also in the correlation chart, we need to get rid of that since it's the response we're looking for.

```r
# create correlation chart 
eth_target_cor_chart <- data.frame(cor(eth_data[-1])) %>%
    select(TARGET_return_nextday)

# order the rows by correlation
colnames(eth_target_cor_chart) <- c('correlation')
correlation_data <- 
    eth_target_cor_chart %>% 
    arrange(desc(correlation)) %>% 
    filter(correlation != 1)


head(correlation_data, 3)
tail(correlation_data, 3)
```

```r
# visualize correlation of inputs with the next-day price
eth_cor_histogram <- 
    correlation_data %>%
    ggplot(aes(x=correlation)) +
    geom_dotplot(binwidth=1/18, fill='white') +
    geom_rect(aes(xmin=-1, xmax=-0.7, ymin=0, ymax=1), fill='red', alpha=0.01) +
    geom_rect(aes(xmin=0.7, xmax=1, ymin=0, ymax=1), fill='blue', alpha=0.01) +
    geom_text(x=-0.85, y=0.8, label='corr < -0.75', size=7) +
    geom_text(x=0.85, y=0.8, label='corr > 0.75', size=7) +
    labs(x='Correlation', y='Density', title='Distribution of correlation of inputs') +
    theme(
        text = element_text(size = 16.5),
        plot.title = element_text(face = "bold"),
        axis.title = element_text(face = "bold"))

eth_cor_histogram
```

Although there's no absolute standard, a correlation $corr>0.75$ or $corr<-0.75$ is usually considered strong. Unfortunately most of our input variables has low correlation level with the response. 
> Consider the price prediction for cryptocurrency being a million-dollar project, nice-and-easy predictors that are strongly correlated to the change in price would have never existed. 

```r
options(repr.plot.width = 10, repr.plot.height = 6)

eth_sample_plot<- eth_data %>%
    ggplot(aes(x = open_interest, y = TARGET_return_nextday))+
    geom_point()+
    geom_smooth(aes(x = open_interest, y = TARGET_return_nextday),method = "lm")+
    xlab("Open interest") +
    ylab("Next-day price ROC") +
    ggtitle("Sample Scatterplot and Estimated SLR with open interest") +
    theme(
        text = element_text(size = 16.5),
        plot.title = element_text(face = "bold"),
        axis.title = element_text(face = "bold")
    )

eth_sample_plot
```

Our plot seems promising already with only one variable. However, notice that as we're targeting at price here, we'd want our prediction to have a relatively small confidence interval so that it wouldn't contain previous day's price too often (in that case our prediction of ETH's gain / loss rate would be inconclusive). That's why we still need to put more variables into consider when conducting the model.

Below we include `business_vol` in addition to `open_interest` to construct a two-variable MLR model, as to check how that fits in a 3-dimension coordinate.

```r
# MLR model
MLR_add_sample <- lm(TARGET_return_nextday~business_vol+open_interest, data=eth_data)

tidy(MLR_add_sample, conf.int=TRUE) %>% mutate_if(is.numeric, round, 2)
```

```r
# scatterplot
s3d <- scatterplot3d(x=eth_data$business_vol, y=eth_data$open_interest, z=eth_data$price, pch=19, type="p", color="black",
                     main="Sample 2-variable MLR addictive model", grid = TRUE, box = FALSE,  
                     mar = c(2.5, 2.5, 2, 1.5), angle = 55, xlab='Business volume', ylab='Open interest', zlab='Price ROC')

s3d$plane3d(MLR_add_sample, draw_polygon = TRUE, draw_lines = TRUE, polygon_args = list(col = rgb(0.8,0.8,0.8,0.8)))
```

Note that most observations lies under the regression plane. The inequal distribution of scatters possibly implies a non-linear association. Therefore, the MLR model we just constructed could be one approach of estimation, but may not be the best fit of our dataset. 

### Splitting training set and testing set
- The dependency in dataset we used for model construction and inference may result in significant errors. Therefore we need to split our original dataset into training set `training_eth` and testing set `testing_eth`.
- As a rule of thumb, we use the $75:25$ split.

```r
set.seed(2022)

eth_data$ID <- 1:nrow(eth_data)

training_eth <- sample_n(eth_data, size = nrow(eth_data) * 0.75,
  replace = FALSE
)

testing_eth <- anti_join(eth_data,
  training_eth,
  by = "ID"
)

training_eth <- training_eth %>% select(-ID)
testing_eth <- testing_eth %>% select(-ID)


head(training_eth, 2)
cat('Number of rows in the training set:', nrow(training_eth))

head(testing_eth, 2)
cat('Number of rows in the testing set:', nrow(testing_eth))
```

### Selecting the optimum model
- At this stage, we first use different selection methods to obtain most efficient combination of inputs at `training_eth`. The methods includes:
    - Full MLR addiction model
    - Forward selection with minimal $C_p$
    - Backward selection with minimal $C_p$
    - LASSO with minimal MSE
- Then we construct a MLR model for each set of selected input variables, and use the model to predict at `testing_eth`.
- Finally, we compare the performance of our models and choose the most suitable one.

#### 1. Full MLR addiction model
First, we construct a MLR addiction model with full input variables, and use the model to predicts the testing set `testing_eth`.

```r
# construct addition model
MLR_FAM <- lm(TARGET_return_nextday~., data=training_eth)
head(tidy(MLR_FAM, conf.int = TRUE))
cat('Number of coefficients:', nrow(tidy(MLR_FAM)))
```

```r
# predict using testing set and addition model
predicted_with_fam <- predict(MLR_FAM, testing_eth)
```

As we studied in previous section, most input variables have low correlation with our response. Therefore, a more effective MLR model should only keeps determinative inputs. Meantime, we should eliminate less-effective variables for reducing "noises" in our model. This is where **forward selection** and **backward selection** could be implemented. 

#### 2. Forward selection

```r
# run forward selection algorithm
eth_MLR_forward_sel <- regsubsets(
    x = TARGET_return_nextday~.,
    data = training_eth, 
    nvmax = ncol(eth_data), 
    method = "forward",
 )

fwd_summary <- summary(eth_MLR_forward_sel)
```

```r
cat('Selection # with minimum C_p:', which.min(fwd_summary$cp))
```

```r
# visualization
plot(
    fwd_summary$cp ,
    xlab =" Number of Variables", 
    ylab =" Cp",
    type="l",
    main = "Forward Selection Cp vs. #variables")
points (8, fwd_summary$cp [8] , col ="red", cex =2, pch =20)
```

We found that the combination of 8 input variables produces smallest $C_p$ with forward selection algorithm. The selected inputs include: `open_interest_top100`, `open_interest_top10`, `short_margin_call`, `open_interest`, `interest`, `outflow_maker`, `net`. So we can construct a MLR addiction model using these input variables and predict the testing set.

```r
# predict using testing set and forward selection model
forward_model <- 
    lm(TARGET_return_nextday~open_interest_top100 +
       open_interest_top10 +
       short_margin_call +
       open_interest +
       interest +
       outflow_maker +
       net +
       transaction_vol,
    testing_eth)

predicted_with_forwardSelection <- predict(forward_model, testing_eth)
```

#### 3. Backward selection

```r
# run backward selection algorithm
eth_MLR_backward_sel <- regsubsets(
   x = TARGET_return_nextday~.,
   data = training_eth, 
    nvmax = ncol(eth_data), 
    method = "backward",
)

bwd_summary <- summary(eth_MLR_backward_sel)
```

```r
cat('Selection # with minimum C_p:', which.min(bwd_summary$cp))
```

```r
# visualization
plot(
    bwd_summary$cp ,
    xlab =" Number of Variables ", 
    ylab =" Cp",
    type="l",
    main = "Backward Selection Cp vs. #variables")
points (8, bwd_summary$cp [8] , col ="red", cex =2, pch =20)
```

Similar to the forward selection we did previously, backward selection also gives a combination of 8 inputs for the smallest value of $C_p$, including `open_interest_top50`, `open_interest_top30`, `short_margin_call`, `open_interest`, `interest`, `outflow_maker`, `net`, `transaction_vol`. Then we construct a MLR addiction model using the selected inputs, and use the model to predict the testing set.

```r
# predict using testing set and backward selection model
backward_model <- 
    lm(TARGET_return_nextday~open_interest_top50 +
        open_interest_top30 + 
        short_margin_call +
        open_interest +
        interest +
        outflow_maker +
        net +
        transaction_vol, 
    testing_eth)

predicted_with_backwardSelection <- predict(backward_model, testing_eth)
```

Sometimes one input variable could play a more important role in prediction than another. So we need to re-weight the coefficients using shrinkage algorithm **LASSO**. Since we construct the model with training set and predict the testing set, we are performing **post LASSO**.

#### 4. LASSO

```r
# construct matrices
TARGET_return_nextday <- training_eth$TARGET_return_nextday
x_vars <- model.matrix(object = TARGET_return_nextday~.,
                      data = training_eth)[,-1]
```

```r
# run LASSO shrinkage algorithm
eth_cv_lambda_lasso <- cv.glmnet(
    x = x_vars,
    y = TARGET_return_nextday,
    alpha = 1)
eth_cv_lambda_lasso$lambda.min
```

```r
# visualization 
plot(eth_cv_lambda_lasso)
plot(eth_cv_lambda_lasso$glmnet.fit, "lambda")
abline(v = log(eth_cv_lambda_lasso$lambda.min), col = "red", lwd = 3, lty = 2)
```

```r
# predict using testing set and LASSO model
lasso_pred <- predict(eth_cv_lambda_lasso, 
                      newx = as.matrix(testing_eth %>% select(-TARGET_return_nextday)))
```

### Model evaluation
At this stage, we aim to select the most suitable model. The model qualification mainly depends on $RMSE$, but we are also interested in different models' capability in explaining and predicting market trends. The methods include:
- First, we compares the RMSE of the predicted results from each model with respect to the actuals, and choose the one with smallest RMSE as the most suitable model.
- Next, we want to test how well our models perform in predicting and explaining general market trends.
- To go even further, we can examine the models with significant market activities.

#### 1. RMSE

```r
model_rmse <- tibble(
    Model = c('Full OLS', 'Forward selection', 'Backward selection', 'LASSO'),
    R_MSE = c(rmse(preds = predicted_with_fam, 
                   actuals = testing_eth$TARGET_return_nextday), 
             rmse(preds = predicted_with_forwardSelection, 
                 actuals = testing_eth$TARGET_return_nextday), 
             rmse(preds = predicted_with_backwardSelection, 
                 actuals = testing_eth$TARGET_return_nextday),
             rmse(preds = lasso_pred, 
                 actuals = testing_eth$TARGET_return_nextday))
)

model_rmse
```

By comparing $RMSE$ of each model, we found the prediction with **forward selection** has the minimized error with $RMSE\approx 0.054$, followed by backward selection and LASSO with minor differences. Full OLS model produces the highest $RMSE$.

#### 2. Market trending
Since our models aim for good performance in real-time market, we are interested in the predictions for market trends. Specifically, we checked whether the predicted direction of price change matches the reality by checking if their products are positive. Furthermore, we neglected insignificant market activities defined by any changes smaller than $0.015$. 
> Notice the threshold for determining insignificant changes $0.015$ is subject to different purposes. The threshold shall be lowered for short-term traders, and the accuracy in prediction will drop correspondingly.

```r
# MLR addiction model
predicted_traget_comparison <- cbind(testing_eth, predicted_with_fam) %>%
    select(TARGET_return_nextday, predicted_with_fam) %>%
    filter(TARGET_return_nextday * predicted_with_fam > 0
           | (((-0.015 <= TARGET_return_nextday) & (TARGET_return_nextday <= 0.015))
              & ((-0.015 <= predicted_with_fam) & (predicted_with_fam <= 0.015))))

fam_correct_prediction_rate <- nrow(predicted_traget_comparison) / length(predicted_with_fam)
```

```r
cat("Full Addition Model correctly predicts ", round(fam_correct_prediction_rate, 2) * 100, "% trending in testing dataset")
```

```r
# forward selection model
predicted_forward_comparison <- cbind(testing_eth, predicted_with_forwardSelection) %>%
    select(TARGET_return_nextday, predicted_with_forwardSelection) %>%
    filter(TARGET_return_nextday * predicted_with_forwardSelection > 0
           | ( ((-0.015 <= TARGET_return_nextday) & (TARGET_return_nextday <= 0.015))
              & ((-0.015 <= predicted_with_forwardSelection) & (predicted_with_forwardSelection <= 0.015))))

forward_correct_prediction_rate <- nrow(predicted_forward_comparison) / length(predicted_with_forwardSelection)
```

```r
cat("Forward Selection correctly predicts", round(forward_correct_prediction_rate, 2) * 100, "% trending in testing dataset")
```

```r
# backward selection model
predicted_backward_comparison <- cbind(testing_eth, predicted_with_backwardSelection) %>%
    select(TARGET_return_nextday, predicted_with_backwardSelection) %>%
    filter(TARGET_return_nextday * predicted_with_backwardSelection > 0
          | ( ((-0.015 <= TARGET_return_nextday) & (TARGET_return_nextday <= 0.015))
              & ((-0.015 <= predicted_with_backwardSelection) & (predicted_with_backwardSelection <= 0.015))))

backward_correct_prediction_rate <- nrow(predicted_backward_comparison) / length(predicted_with_backwardSelection)
```

```r
cat("Backward Selection correctly predicts", round(backward_correct_prediction_rate, 2) * 100, "% trending in testing dataset")
```

We found the forward selection model capable in predicting the most trends, with a 2% difference to the backward model. Full addiction model still has the worst performance. 

#### 3. Significant market activities
Next, we apply our models under significant market changes by only considering rates higher than $0.04$. Again the figure we used is subjective and changeable. 

```r
# identify "significant" activities
testing_eth_w_roctype <- testing_eth %>%
    mutate(roc_type = if_else(abs(TARGET_return_nextday) > 0.04, "Significant", "Insignificant"))
```

```r
# filter "significant" activities
testing_eth_significant <- filter(testing_eth_w_roctype, roc_type == "Significant")
```

```r
# forward selection model
predicted_with_forwardSelection_sig <- predict(forward_model, testing_eth_significant)
predicted_forward_comparison_sig <- cbind(testing_eth_significant, predicted_with_forwardSelection_sig) %>%
    select(TARGET_return_nextday, predicted_with_forwardSelection_sig)

sig_prop_forward <- nrow(filter(predicted_forward_comparison_sig,
    abs(predicted_with_forwardSelection_sig) > 0.04))

cat(sig_prop_forward, 'successful prediction out of', nrow(testing_eth_significant), 'significant changes')
```

```r
# backward selection model
predicted_with_backwardSelection_sig <- predict(backward_model, testing_eth_significant)
predicted_backward_comparison_sig <- cbind(testing_eth_significant, predicted_with_backwardSelection_sig) %>%
    select(TARGET_return_nextday, predicted_with_backwardSelection_sig)

sig_prop_backward <- nrow(filter(predicted_backward_comparison_sig,
    abs(predicted_with_backwardSelection_sig) > 0.04))

cat(sig_prop_backward, 'successful prediction out of', nrow(testing_eth_significant), 'significant changes')
```

Our models perform poorly under significant market activities, which gives us spaces for future improvements. Nevertheless, predicting the surges and plummets of prices could be a challenge even for the most comprehensive prediction model. 

## Discussion
---

### Findings
- In this project report, we looked over the data of the crypto-currency ETH over the past year and trained four different models using various featured selection methods. 
- Then we compared $RMSE$ and performance in predicting market trendings for these models. And we found:
    - the model generated with *forward selection* gives the smallest $RMSE$, 
    - whereas the *backward selection* model predicts the most upcoming trends ($73\%$ of the trends).
    
### Implications
- For the findings we have throughout the study, we conclude that, subject to different trading styles and demands, both forward-selection model and backward-selection model could be the most suitable in specific context.

- Our study explored an approach to quantitatively predict the highly-fluctuated cryptocurrency market by statistical models. Consider the great uncertainty of the crypto market, our result is accptable. 

- Also, we found the model's shortcomings at predicting upcoming significant changes. This raised our interests as well as further research questions (that we will discuss in the following section). This also shows us one direction on how to improve our current model.

### Improvements and Future studies

- Considering how fast the world of crypto-currency is changing, it could be possible that a model based on data collected over one year ago might be irrelevant to the current market. We could try to shorten the period of time from which we collect our data, in this case, having a smaller dataset might actually be beneficial to our model.

- Currently, our model isn't very good at predicting upcoming significant changes in the market. This could potentially improve if we would increase the weight of the data points with a significant rate of change when training the model. However, that could very possibly work both ways and instead of dragging the prediction on trends worse than our current model.

- Continuing on the former idea, given that the predicted values of the forward-selection model are highly conservative compared to the actual market, we wonder if it is because we have somehow scaled our response variable unintentionally when training our model. We should look into that and see if things would change if we can find a mathematically logical way to scale the predictions to be more aggressive.

- Instead of just simply dividing the response variable (rate of change) into two classes: insignificant and significant, we could try to cluster them into several classes, and train a classification model instead. That would definitely be more targeted and flexible when we want to change the trading strategy from long-term investment to short-term investment.

- Continuing on the former point, for the model to have the potential of being able to predict a longer-term trend, we could change our response variable to be the rate of change of the next week or even longer. However, it's very likely that those results would be even harder to predict.
